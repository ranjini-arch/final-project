## Data Cleaning and Pre-processing

- Handle missing values

   Missing values are empty or null data present in the dataset. These values are checked using isnull() and handled by replacing them with suitable values or removing them to improve data quality.

- Handle Duplicates

   Duplicate records are repeated rows in the dataset. These duplicates can affect analysis accuracy, so they are identified using duplicated() and removed using drop_duplicates().

- Rename Columns or Drop Unwanted Columns

  Some column names may be unclear or unnecessary for analysis. Renaming columns improves readability, while unwanted columns like id can be removed to simplify the dataset.

- Treat outliers

  Outliers are extreme or unusual values that differ greatly from other data points. They are identified using boxplots and treated if necessary to avoid incorrect analysis results.

- Check skewness and apply transformations

  Skewness shows whether data distribution is balanced or uneven. Highly skewed data can affect analysis, so transformations like log transformation are applied to normalize the data.

- Convert data types

  Data type conversion ensures columns are stored in the correct format such as integer, float, or datetime. This helps perform accurate calculations and analysis.

- Feature transformations

  Feature transformation creates new useful columns from existing data. For example, extracting year and month from the date column helps analyze seasonal and yearly trends in PMFBY data.

 Outlier Detection and Removal

```python
#  Outlier Detection and Removal
#  Detect Outliers in premium_amount
Q1 = df['premium_amount'].quantile(0.25)
Q3 = df['premium_amount'].quantile(0.75)
IQR = Q3 - Q1
LB = Q1 - 1.5 * IQR
UB = Q3 + 1.5 * IQR

Q1,Q3,IQR,LB,UB
```

Output

* Outliers in premium_amount are identified successfully.
* A few states may have unusually high premium amounts compared to others.
* These extreme values can be reviewed, treated, or removed if necessary.
* Detecting outliers helps create a more reliable PMFBY analysis and dashboard.

```python
outlier_col = ['loanee',
               'non_loan_former',
               'insurance_units',
               'premium_amount']

for col in outlier_col:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    LB = Q1 - 1.5 * IQR
    UB = Q3 + 1.5 * IQR

    df = df[(df[col] >= LB) & (df[col] <= UB)]

    print(f"Shape after removing outliers from {col}:", df.shape)
```
